If we create two ricker wavelets with the same central frequency but generte different sequence lengths, then we get different discrete sample sequences representing that same continuous wavelet.

```python
import matplotlib
%matplotlib inline
import matplotlib.pylab as plt
import os, sys
import bruges
import numpy as np

plt.figure(figsize=(20,5))

#  4ms sampling
Fs = 250
dt = 1/Fs

ricker = bruges.filters.wavelets.ricker
# 30Hz Ricker wavelet odd num samples
w_odd = ricker(0.072, dt, 30)

plt.subplot(121)
plt.stem(dt*np.array(range(0,len(w_odd))), w_odd)
plt.plot(dt*np.array(range(0,len(w_odd))), w_odd)
plt.xlabel('Time (s)')
plt.axis('tight')
plt.title('30Hz Ricker')

# 30Hz Ricker wavelet even num samples
w_even = ricker(0.076, dt, 30)

plt.subplot(122)
plt.stem(dt*np.array(range(0,len(w_even))), w_even)
plt.plot(dt*np.array(range(0,len(w_even))), w_even)
plt.xlabel('Time (s)')
plt.axis('tight')
plt.title('30Hz Ricker')
```

if we look at the spectra, it does not affect the frequency but does increase the ampltiude of the wavelet

```python
N = 64
f = Fs*np.fft.fftshift(np.fft.fftfreq(N))
W_odd = np.fft.fftshift(np.fft.fft(w_odd, N))
W_even = np.fft.fftshift(np.fft.fft(w_even, N))
```

```python
plt.figure(figsize=(20,5))
    
plt.subplot(121)
plt.plot(f, np.abs(W_odd))
plt.plot(f, np.abs(W_even),'r')
plt.title('Amplitude Spectra')
plt.xlabel('Frequency (Hz)')

plt.subplot(122)
plt.plot(f, np.angle(W_odd))
plt.plot(f, np.angle(W_even),'r')
plt.title('Phase Spectra')
plt.xlabel('Frequency (Hz)')
```

```python
print('Peak Frequency (Odd)', abs(f[W_odd.argmax(axis=0)]), 'Hz')
print('Peak Frequency (Even)', abs(f[W_even.argmax(axis=0)]), 'Hz')
```
